```python
import pandas as pd
import random
```

```python
employees = {
    "name": ["Alice Johnson", "Joe Smith", "Carmen Ortiz", "Deepak Singh", "Elena Petrova", 
             "Frank Wright", "Grace Lee", "Henry Adams", "Ivy Chen", "Jack Smith", 
             "Kathy Brown", "Leo White", "Mia Green", "Nina Black", "Oscar Blue","Ramesh"],
    "age": [25, 30, 30, 25, 35, 
            40, 25, 35, 30, 25, 
            45, 35, 30, 35, 40,22],
    "salary": [82000, 115000, 105000, 89000, 97000, 
               110000, 78000, 95000, 102000, 88000, 
               120000, 90000, 90000, 94000, 100000,90000],
    "performance_score": [random.randint(1, 100) for _ in range(16)]
}

df = pd.DataFrame(employees)
```

```python
df.sort_values(by=["salary"],ascending=[True],inplace=True,axis=0)
print(df)
```

```python
avg_salary = df['salary'].mean()
print(f"Average Salary: {avg_salary}")
```

```python
grouped = df.groupby(["age","performance_score"])["salary"].median()
print(grouped)
```
