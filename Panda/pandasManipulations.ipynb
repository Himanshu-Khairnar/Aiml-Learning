### Handling Missing Data

##### fillna(value, inplace=True)
Fills missing values with the specified value. When `inplace=True`, modifies the DataFrame directly.

##### dropna(axis=0, inplace=True)
##### interpolate(method='linear', axis=0, inplace=True)
Fills `NaN` values using an interpolation method. By default, it uses linear interpolation between data points. When `inplace=True`, modifies the DataFrame directly.


```python
# df["age"].fillna(round(df["age"].mean()),inplace=True)
df["salary"].fillna(round(df["salary"].mean()),inplace=True)

df["age"].interpolate(method='linear',axis=0,inplace=True)

# df.dropna(axis=1,inplace=True)
print(df)
```
